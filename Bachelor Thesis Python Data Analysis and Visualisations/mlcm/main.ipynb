```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os

for file in os.listdir("csv_files"):
    x_0=[]
    y_0=[]
    x_4=[]
    y_4=[]
    x_16=[]
    y_16=[]
    lower_limits_0=[]
    higher_limits_0=[]
    lower_limits_4=[]
    higher_limits_4=[]
    lower_limits_16=[]
    higher_limits_16=[]
   
    result=pd.read_csv("csv_files"+"/"+file)
    for i in range(len(result["scale"])):
        if result["sf"][i]== 0.5:
            x_0.append(result["contrast"][i])
            lower_limits_0.append(result["scale"][i]-result["low"][i])
            y_0.append(result["scale"][i])
            higher_limits_0.append(result["high"][i]-result["scale"][i])  
        if result["sf"][i]== 4.0:
            x_4.append(result["contrast"][i])
            lower_limits_4.append(result["scale"][i]-result["low"][i])
            y_4.append(result["scale"][i])
            higher_limits_4.append(result["high"][i]-result["scale"][i])
        if result["sf"][i]== 16.0:
            x_16.append(result["contrast"][i])
            lower_limits_16.append(result["scale"][i]-result["low"][i])
            y_16.append(result["scale"][i])
            higher_limits_16.append(result["high"][i]-result["scale"][i])
    lower_limits_0=np.array(lower_limits_0)
    higher_limits_0=np.array(higher_limits_0)
    lower_limits_4=np.array(lower_limits_4)
    higher_limits_4=np.array(higher_limits_4)
    lower_limits_16=np.array(lower_limits_16)
    higher_limits_16=np.array(higher_limits_16)
    ax =plt.errorbar(x_0,y_0, yerr=[lower_limits_0,higher_limits_0],c="#c7e9b4", label=f"0.5 cpd",ecolor="#c7e9b4", capsize=5)
    ax1=plt.errorbar(x_4,y_4,yerr= [lower_limits_4,higher_limits_4],c="#41b6c4",label=f"4 cpd",ecolor="#41b6c4",capsize=5)
    ax2=plt.errorbar(x_16,y_16,yerr= [lower_limits_16,higher_limits_16],c="#253494",label=f"16 cpd",ecolor="#253494",capsize=5)
    plt.ylim((0, 13))
    plt.xlabel("Contrast x")
    plt.ylabel("Perceived contrast $\Psi$(x)")
    plt.title(f"{file[0:2].upper()}")
    sns.despine()
    plt.legend()
    plt.show()
    print(file[0:2])
    #normalized
    ax =plt.plot(x_0,y_0/np.max(y_0),label=f"0.5 cpd",color="#c7e9b4",marker="o")
    ax1=plt.errorbar(x_4,y_4/np.max(y_4),label=f"4 cpd",color="#41b6c4",marker="o")
    ax2=plt.errorbar(x_16,y_16/np.max(y_16),label=f"16 cpd",color="#253494",marker="o")
    plt.xlabel("Contrast x")
    plt.ylabel("Perceived contrast $\Psi$(x)")
    plt.title(f"{file[0:2].upper()}")
    sns.despine()
    plt.legend()
    plt.show()
    
    

```
